```python
infile="03a_input.txt"
```

```python
with open(infile) as f:
    lines=f.readlines()
```

```python
lines=[l.replace("\n","") for l in lines]
lines
```

```python
numrows=len(lines)
numcols=len(lines[0])
print("rows:",numrows,", cols:",numcols)
```

```python
def count_trees(lines, dcol,drow):
    trees=0
    
    numrows=len(lines)
    numcols=len(lines[0])

    # current position
    col=0
    row=0
    
    # go skiing
    for idx in range(numrows):
        if row>=numrows:
            break

        # position in repeated pattern (modulo)
        colmod=col%numcols

        # count crosses=trees
        item=lines[row][colmod]
        if item=="#":
            trees+=1

        # update skier position    
        row=row+drow
        col=col+dcol
        
    return trees
```

```python
# (A) tobogganning slope
dcol=3
drow=1

trees=count_trees(lines,dcol,drow)
print("Trees:",trees)
```

#### Part B

```python
t1=count_trees(lines,1,1)
t2=count_trees(lines,3,1)
t3=count_trees(lines,5,1)
t4=count_trees(lines,7,1)
t5=count_trees(lines,1,2)
```
