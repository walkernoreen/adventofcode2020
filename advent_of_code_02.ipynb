#### Part A

```python
import numpy as np
import pandas as pd
```

```python
infile="02a_input.txt"
```

```python
df=pd.read_csv(infile,sep=" ",header=None,names=["range","letter","pwd"])
df
```

```python
# clean up table: remove the ":", and split min and max count into two separate columns
df["letter"]=df["letter"].apply(lambda x: x.replace(":","")) #or: df["letter"].str.replace(":","")
df["start"]=df["range"].apply(lambda x: int(x.split("-")[0]))
df["stop"]=df["range"].apply(lambda x: int(x.split("-")[-1]))
df.head()
```

```python
# count the number of letter occurences and whether they are within the limits
df['count'] = df.apply(lambda x:  x['pwd'].count(x['letter']), axis=1)
df['correct']=df.apply(lambda x: x['count']>=x['start'] and x['count']<=x['stop'],axis=1)
df
```

```python
# count the number of correct pwds ("true")
df["correct"].value_counts()
```
