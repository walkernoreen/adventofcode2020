```python
import re
```

```python
infile="19a_input.txt"
#infile="dummy.txt"
```

```python
with open(infile) as f:
    lines=f.readlines()
```

```python
instructions=[]
samples=[]

reading_instr=True
for line in lines:
    if line=="\n":
        reading_instr=False
        continue
        
    if reading_instr:
        line=line.strip()
        instructions.append(line.replace('"',''))
        
    else:
        samples.append(line.strip())
```

```python
instr=[i.split(": ") for i in instructions]
```

```python
instrdict=dict(instr)
instrdict
```

```python
samples
```

```python
# replace each number with the condition and do so recursivle - create a giant regex
cond=instrdict["0"]
match=re.search(r"\d+",cond)
while match is not None:
    span=match.span()
    newrule=instrdict[match.group()]
    if newrule!="a" and newrule!="b":
        newrule="("+newrule+")"
        #newrule=newrule.replace("|",")|(")
        cond=cond[0:span[0]]+newrule+cond[span[1]:] # replace in string
    else:
        cond=cond[0:span[0]]+newrule+cond[span[1]:] # replace in string
        
    #print(cond)
    match=re.search(r"\d+",cond)    
```

```python
condfinal=cond.replace(" ","")
condfinal
```

```python
countvalid=0
for idx,sample in enumerate(samples):
    if re.fullmatch(condfinal,sample) is not None:
        countvalid+=1
        print(idx,sample,"is valid")
```

```python
#final result
countvalid
```
